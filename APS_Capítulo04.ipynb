## 1. Metodologias Ágeis e Gestão de Projetos

### **Conceito:**
Metodologias Ágeis são abordagens iterativas e incrementais que priorizam a entrega contínua de valor para o cliente, adaptando-se às mudanças durante o desenvolvimento de um projeto.

### **Exemplo:**
Imagine uma equipe desenvolvendo um aplicativo para controle de despesas pessoais. Em vez de planejar todo o projeto de uma só vez, as funcionalidades são entregues em etapas, como "Cadastro de Despesas", "Gráficos de Análise" e "Relatórios Mensais".

## 2. Scrum: Papéis, Eventos e Artefatos

### **Conceito:**
Scrum é uma metodologia ágil que organiza o trabalho em sprints (ciclos curtos de desenvolvimento) e se baseia em papéis, eventos e artefatos.

### **Papéis:**
- **Product Owner:** Responsável por maximizar o valor do produto e garantir que o backlog esteja priorizado e alinhado com as necessidades do cliente. Atua como um ponto de conexão entre o cliente e o time.
- **Scrum Master:** Facilita o processo Scrum, remove impedimentos e assegura que a equipe siga os princípios ágeis. Não é um gestor, mas um facilitador.
- **Time de Desenvolvimento:** Composto por profissionais multidisciplinares que trabalham para entregar incrementos funcionais do produto ao final de cada sprint.

### **Eventos:**
1. **Planejamento da Sprint:**
   - Encontro onde o time define o objetivo da sprint e seleciona itens do backlog para trabalhar.
   - Os itens são detalhados em tarefas específicas para serem concluídas durante o ciclo.

2. **Reuniões Diárias (Daily Scrum):**
   - Reunião rápida (15 minutos) realizada diariamente para que o time sincronize atividades, identifique impedimentos e ajuste o plano de trabalho.

3. **Revisão da Sprint:**
   - Realizada ao final de cada sprint, onde o time apresenta o incremento do produto aos stakeholders e coleta feedback.

4. **Retrospectiva da Sprint:**
   - O time avalia o processo de trabalho para identificar o que funcionou bem, o que pode ser melhorado e implementações para a próxima sprint.

### **Artefatos:**
1. **Backlog do Produto:**
   - Lista priorizada de tudo que precisa ser desenvolvido no produto.
   - Mantido pelo Product Owner e atualizado continuamente.

2. **Backlog da Sprint:**
   - Subconjunto do backlog do produto que o time se compromete a concluir durante a sprint.

3. **Incremento:**
   - Resultado do trabalho concluído durante a sprint, que deve ser um produto utilizável e que atenda a critérios de "pronto".

### **Exemplo:**
Uma equipe está desenvolvendo um sistema de gestão acadêmica. O Product Owner prioriza funcionalidades como "Cadastro de Alunos" e "Relatórios de Notas". O Time de Desenvolvimento trabalha em sprints de 2 semanas para entregar incrementos do sistema. Durante a Retrospectiva, a equipe identifica melhorias para o próximo sprint.

   - Criar um backlog do produto com 5 itens.
   - Planejar a primeira sprint.
4. Cada grupo apresenta seu planejamento e backlog para a turma.

## 3. Kanban e Gestão Visual

### **Conceito:**
Kanban é uma metodologia visual para gerenciar fluxos de trabalho. O trabalho é representado em cartões organizados em colunas que indicam etapas do processo (ex.: A Fazer, Fazendo, Feito).

- **Princípios:** Visualize o fluxo de trabalho, limite o trabalho em progresso e melhore continuamente.

### **Exemplo:**
Uma equipe está desenvolvendo um site e utiliza um quadro Kanban:
- **Coluna "A Fazer"**: Design da página inicial.
- **Coluna "Fazendo"**: Implementação do formulário de contato.
- **Coluna "Feito"**: Configuração do banco de dados.

### **Atividade Prática:**
1. Usando um quadro físico ou ferramenta online (ex.: Trello), crie um quadro Kanban para um projeto fictício como "Organização de um Evento Acadêmico".
2. Liste tarefas como "Convidar Palestrantes", "Criar Campanha de Divulgação" e "Preparar Material de Apoio".
3. Movimente as tarefas pelo quadro conforme as etapas são realizadas.

## 4. Integração das Práticas Ágeis com a Modelagem de Sistemas

### **Conceito:**
As práticas ágeis podem ser integradas à modelagem de sistemas, permitindo ajustes contínuos à medida que o projeto evolui. Exemplos incluem a atualização de diagramas UML durante as sprints.

### **Exemplo:**
Em um projeto de desenvolvimento de um sistema de agendamento de consultas médicas, a equipe começa com um diagrama de classes simples para "Paciente" e "Consulta". Após feedback do cliente, novos atributos e relações são adicionados em cada sprint.

   - Simular a evolução do sistema em 2 sprints, ajustando o diagrama a cada iteração.
