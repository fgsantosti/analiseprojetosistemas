### **Projetos Reais para Simular a Metodologia Scrum em Sala de Aula**

Cada projeto abaixo foi pensado para ser desenvolvido em equipe (três participantes por grupo) e exige que os alunos sigam as etapas da metodologia Scrum, como **planejamento do backlog**, **definição de sprints**, **daily meetings**, **review** e **retrospective**.

Os projetos possuem escopo suficiente para simular um ambiente de trabalho ágil em uma aula prática.

---

### **1. Sistema de Gestão de Estoque**

- Cadastro de produtos (nome, quantidade, preço, etc.).
- Consulta de estoque disponível.
- Relatório de itens abaixo do estoque mínimo.

### **Estrutura para Implementação de Cada Projeto com Scrum**
1. **Reunião Inicial (Planning Meeting)**:
   - Os alunos criam o **Product Backlog** do projeto, listando todas as funcionalidades a serem implementadas.
   - Priorizar os itens do backlog e definir os objetivos para a primeira sprint.

2. **Sprints (Iterações)**:
   - Cada sprint deve durar entre **15 e 20 minutos** (simulação compactada).
   - Os alunos desenvolvem incrementos do projeto em cada sprint.

3. **Daily Meetings**:
   - A cada 10 minutos, os alunos fazem uma reunião rápida (simulada) para compartilhar o progresso e identificar impedimentos.

4. **Sprint Review**:
   - Ao final de cada sprint, o grupo apresenta o incremento do projeto para a equipe ou para o professor.

5. **Sprint Retrospective**:
   - Após a apresentação, discutem o que funcionou bem, o que precisa melhorar e ajustam os processos para a próxima sprint.

---

### **Critérios de Avaliação**
- **Organização do Backlog (20%)**:
  - Divisão clara e coerente das tarefas.
- **Participação e Comunicação (20%)**:
  - Todos os membros do grupo devem participar e colaborar.
- **Entrega dos Incrementos (40%)**:
  - Desenvolvimento de funcionalidades alinhadas ao backlog e às prioridades.
- **Aplicação da Metodologia Scrum (20%)**:
  - Uso correto das cerimônias (planning, daily, review e retrospective).
